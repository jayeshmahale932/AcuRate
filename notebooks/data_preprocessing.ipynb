# Handling Missing Values

```python
# Drop columns with ≥50% missing values
null_percent = df.isnull().mean() * 100
high_null_cols = null_percent[null_percent >= 50]
print("No. of colums with >=50% null values:", len(high_null_cols))

df = df.drop(columns=high_null_cols.index)
print(f"Remaining columns: {df.shape[1]}")
```

| Column                                                                                         | Reason                                                        |
| ---------------------------------------------------------------------------------------------- | ------------------------------------------------------------- |
| `id`                                                                                           | Just a unique identifier                                      |
| `emp_title`                                                                                    | Free-text, unstructured job title                             |
| `title`                                                                                        | User-provided loan title (redundant, high cardinality)        |
| `url`                                                                                          | Obsolete/irrelevant                                           |
| `zip_code`                                                                                     | Partially masked → low value, high leakage risk               |
| `addr_state`                                                                                   | Could bias the model geographically (optional to drop)        |
| `earliest_cr_line`                                                                             | Date as string → better to convert to derived feature         |
| `application_type`                                                                             | Most entries are "Individual", low variance                   |
| `policy_code`                                                                                  | Constant = 1, no variance                                     |
| `issue_d`, `last_pymnt_d`, `last_credit_pull_d`, `next_pymnt_d`                                | Temporal leakage: not available at loan origination           |
| `pymnt_plan`                                                                                   | Almost always 'n', low variance                               |
| `out_prncp`, `out_prncp_inv`, `total_pymnt`, `recoveries`, `collection_recovery_fee`           | Post-loan info → data leakage                                 |
| `loan_status`, `funded_amnt`, `funded_amnt_inv`                                                | Correlated with outcome or future status                      |
| `member_id`                                                                                    | 100% null, already dropped                                    |
| `total_pymnt_inv`, `total_rec_prncp`, `total_rec_int`, `total_rec_late_fee`, `last_pymnt_amnt` | Post-loan → Data leakage                                      |
| `last_fico_range_high`, `last_fico_range_low`                                                  | After disbursement → leakage                                  |
| `collections_12_mths_ex_med`                                                                   | Almost always 0 (check distribution)                          |
| `acc_now_delinq`, `delinq_amnt`, `num_tl_120dpd_2m`, `num_tl_30dpd`, `num_tl_90g_dpd_24m`      | Too sparse or duplicate of better vars                        |
| `chargeoff_within_12_mths`, `tax_liens`                                                        | Too sparse or low signal (EDA confirms)                       |
| `disbursement_method`                                                                          | Almost always "Cash"                                          |
| `hardship_flag`, `debt_settlement_flag`                                                        | Post-loan hardship → target leakage                           |
| `pub_rec_bankruptcies`                                                                         | Can keep if clean, but usually sparsely filled (EDA required) |


```python
# Drop obvious columns
drop_cols = ['id', 'emp_title', 'title', 'url', 'zip_code', 'addr_state', 'earliest_cr_line', 'application_type',
    'policy_code', 'issue_d', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d',
    'pymnt_plan', 'out_prncp', 'out_prncp_inv', 'total_pymnt', 
    'recoveries', 'collection_recovery_fee', 'loan_status', 'funded_amnt', 
    'funded_amnt_inv', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee', 'last_pymnt_amnt', 'last_fico_range_high',
    'last_fico_range_low', 'collections_12_mths_ex_med', 'acc_now_delinq', 'delinq_amnt', 'num_tl_120dpd_2m',
    'num_tl_30dpd', 'num_tl_90g_dpd_24m', 'chargeoff_within_12_mths', 'tax_liens', 
    'disbursement_method', 'hardship_flag', 'debt_settlement_flag', 'pub_rec_bankruptcies', 
    'pub_rec_bankruptcies'
]
df = df.drop(columns=[col for col in drop_cols if col in df.columns])
df.shape
```

```python
missing_info = pd.Series()
for col in df.isna().sum().index:
    if df[col].isna().sum() > 0:
        missing_info[col] = df[col].isna().sum()
print(missing_info)
```

```python
# 1. Drop 33 rows with core nulls
df = df.dropna(subset=['loan_amnt', 'term', 'int_rate'])

# 2. Impute low-null numerical features
num_impute_cols = ['annual_inc', 'dti', 'delinq_2yrs', 'inq_last_6mths', 
                   'open_acc', 'pub_rec', 'total_acc', 'revol_util']
for col in num_impute_cols:
    df[col] = df[col].fillna(df[col].median())

# 3. Impute categorical with 'unknown'
df['emp_length'] = df['emp_length'].fillna('unknown')

# 4. Drop high-missing features
drop_high_nulls = ['open_acc_6m', 'open_il_12m', 'open_il_24m', 'total_bal_il', 
                   'il_util', 'inq_fi', 'total_cu_tl', 'inq_last_12m', 'open_act_il',
                   'max_bal_bc', 'all_util', 'mo_sin_old_il_acct', 'mths_since_rcnt_il',
                   'tot_coll_amt', 'tot_cur_bal', 'total_rev_hi_lim',
                   'acc_open_past_24mths', 'avg_cur_bal', 'bc_util', 'bc_open_to_buy',
                   'mo_sin_rcnt_rev_tl_op', 'mo_sin_old_rev_tl_op', 'mo_sin_rcnt_tl',
                   'mths_since_recent_bc', 'mths_since_recent_inq', 'num_accts_ever_120_pd',
                   'num_actv_bc_tl', 'num_actv_rev_tl', 'num_bc_sats', 'num_bc_tl',
                   'num_il_tl', 'num_op_rev_tl', 'num_rev_accts', 'num_rev_tl_bal_gt_0',
                   'num_sats', 'num_tl_op_past_12m', 'pct_tl_nvr_dlq', 'percent_bc_gt_75',
                   'tot_hi_cred_lim', 'total_bc_limit', 'total_bal_ex_mort',
                   'total_il_high_credit_limit', 'mort_acc']
df.drop(columns=drop_high_nulls, inplace=True)
df.shape
```
